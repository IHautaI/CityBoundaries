```python
import geojson
```

```python
with open('output.json', 'r', encoding='iso 8859-1') as fh:
    data = geojson.load(fh)
```

```python
data.features[0]
```

```python
data.features[0].properties['NAME10']
```

```python
for item in data.features:
    with open(str(item.properties['GEOID10']) + '.json', 'w', encoding='utf-8') as fh:
        fh.write(geojson.dumps(geojson.FeatureCollection([item]), sort_keys=True))
```

```python
names = {}
for item in data.features:
    names[item.properties['NAME10']] = item.properties['GEOID10']
```
